# Uni - Variate analysis
- Uni-variate analysis is the analysis involving a single variable (‘uni’) without considering relationships with other variables. This is the stand-alone analysis of a variable/column without considering any casual relationships with other variables. We will see how a simple uni-variate analysis can help to get more insights into the data.

### Histogram :

```python
plt.hist(data['Item_Visibility'], bins=5)
plt.show()
```

### Hist Plot :

```python
sns.histplot(x='Item_MRP', data=data, kde=True)
```

### Strip Plot :

```python
sns.stripplot(x=data["Item_Visibility"])
```

### Pie Chart :

```python
data['Outlet_Size'].value_counts().plot(kind="pie", autopct="%.2f")
plt.title('Outlet Sizes of Products')
plt.show()
```

- Outlet Size '2' has higher percentage than other Products.

### Box Plot :

```python
sns.boxplot(x=data['Item_Outlet_Sales'])
```

### Removing outliers :

```python
# IQR
Q1 = np.percentile(data['Item_Outlet_Sales'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(data['Item_Outlet_Sales'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", data.shape)
 
# Upper bound
upper = np.where(data['Item_Outlet_Sales'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(data['Item_Outlet_Sales'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)
 
print("New Shape: ", data.shape)
```

```python
sns.boxplot(data['Item_Outlet_Sales'])
```

### Heat Map :

```python
sns.heatmap(data.corr(),annot=True,cmap='RdBu')
plt.show()
```

### CountPlot :
